```python
#add path to dataset
import pandas as pd
import seaborn as sns
import matplotlib
from matplotlib.cbook import boxplot_stats
import matplotlib.pyplot as plt
df = pd.read_csv('', delimiter=';')

%matplotlib inline
```

```python
#plot boxplots for each variable to identify outliers
names = ['landing visitors', 'payment sessions', 'registrations', 'plans', 'payment data', 'access']
fig, axes = plt.subplots(6,1, figsize = (5,30))
for i, t in enumerate(names):
    sns.boxplot(y = 'value', x = 'variable', data = pd.melt(df[[t]]),  width = 0.2,
                 palette = 'pastel', ax = axes[i % 6] )
```

```python
#print outliers nicely for each variable
for name in names:
    outliers = [y for stat in boxplot_stats(df[name]) for y in stat['fliers']]
    if outliers == []:
        print('VARIABLE ' + name + ' - no outliers\n')
    else :
        df_outliers = df[df[name].isin(outliers)][[name, 'Date']] 
        print('VARIABLE ' + name + ' - outliers ' + str(list(df_outliers[name])) 
              + ', on dates ' + str(list(df_outliers['Date'])) + '\n')
```
